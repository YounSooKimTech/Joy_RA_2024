```python
text = """Fly me to the moon
Let me play among the stars
Let me see what spring is like on
A-Jupiter and Mars
In other words, hold my hand
In other words, baby, kiss me
Fill my heart with song and let me sing forevermore
You are all I long for
All I worship and adore
In other words, please be true
In other words, I love you
Fill my heart with song
Let me sing forevermore
You are all I long for, all I worship and adore
In other words, please be true
In other words
In other words
I love you"""
```

```python
from nrclex import NRCLex

emotion = NRCLex(text)

print(emotion.top_emotions, "\n")
print(emotion.raw_emotion_scores, "\n")

print(emotion.affect_list, "\n")
print(emotion.affect_dict, "\n")
```

```python
import pandas as pd

affect_dict = emotion.affect_dict

for key, value in affect_dict.items():
  print(f"{key}: {value}")
```
